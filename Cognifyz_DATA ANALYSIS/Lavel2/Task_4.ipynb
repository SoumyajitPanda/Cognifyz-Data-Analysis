Task: Rastaurant Chains

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Dataset
df = pd.read_csv("../Dataset .csv")
```

```python
# Find restaurant chains
restaurant_count = df["Restaurant Name"].value_counts()

chains = restaurant_count[restaurant_count > 1]

top10 = chains.head(10)
```

```python
# Average Rating
chain_rating = df.groupby("Restaurant Name")["Aggregate rating"].mean()

chain_rating = chain_rating.loc[top10.index]
```

```python
# Average Votes
chain_votes = df.groupby("Restaurant Name")["Votes"].mean()

chain_votes = chain_votes.loc[top10.index]
```

```python
# Result Table
result = pd.DataFrame({
    "Branches": top10,
    "Average Rating": chain_rating,
    "Average Votes": chain_votes
})

print(result)
```

```python
# Branches Chart
plt.figure(figsize=(10,5))
plt.bar(result.index, result["Branches"])
plt.xticks(rotation=90)
plt.title("Top Restaurant Chains")
plt.xlabel("Restaurant Name")
plt.ylabel("Number of Branches")
plt.show()
```

```python
# Rating Chart
plt.figure(figsize=(10,5))
plt.bar(result.index, result["Average Rating"])
plt.xticks(rotation=90)
plt.title("Average Rating of Top Restaurant Chains")
plt.xlabel("Restaurant Name")
plt.ylabel("Average Rating")
plt.show()
```
