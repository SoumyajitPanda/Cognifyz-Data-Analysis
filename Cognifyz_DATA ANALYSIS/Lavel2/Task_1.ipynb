Task: Restaurant Ratings

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Dataset
df = pd.read_csv("../Dataset .csv")
```

```python
# Rating Distribution
rating_count = df["Aggregate rating"].value_counts().sort_index()

print(rating_count)
```

```python
# Histogram
plt.figure(figsize=(8,5))
plt.hist(df["Aggregate rating"], bins=10)
plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
# Most Common Rating
most_common_rating = df["Aggregate rating"].mode()[0]
print("Most Common Rating:", most_common_rating)
```

```python
# Rating Range
bins = [0,1,2,3,4,5]

labels = ["0-1","1-2","2-3","3-4","4-5"]

df["Rating Range"] = pd.cut(
    df["Aggregate rating"],
    bins=bins,
    labels=labels,
    include_lowest=True
)

range_count = df["Rating Range"].value_counts().sort_index()

print("\nRating Range Distribution")
print(range_count)

print("\nMost Common Rating Range:", range_count.idxmax())
```

```python
# Average Votes
average_votes = df["Votes"].mean()

print("\nAverage Votes:", round(average_votes,2))
```

```python
# Bar Chart
plt.figure(figsize=(7,5))
plt.bar(range_count.index.astype(str), range_count.values)
plt.title("Restaurant Rating Range Distribution")
plt.xlabel("Rating Range")
plt.ylabel("Number of Restaurants")
plt.show()
```
