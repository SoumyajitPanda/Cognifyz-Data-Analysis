Task: Price Range Distribution

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Dataset
df = pd.read_csv("../Dataset .csv")
```

```python
# Count restaurants by price range
price_count = df["Price range"].value_counts().sort_index()
```

```python
# Calculate percentage
percentage = (price_count / len(df)) * 100
```

```python
# Create result table
result = pd.DataFrame({
    "Price Range": price_count.index,
    "Restaurant Count": price_count.values,
    "Percentage (%)": percentage.values
})

result["Percentage (%)"] = result["Percentage (%)"].round(2)

print(result)
```

```python
# Most common price range
print("\nMost Common Price Range:", price_count.idxmax())
```

```python
# Bar Chart
plt.figure(figsize=(7,5))
plt.bar(result["Price Range"].astype(str), result["Restaurant Count"])

plt.title("Distribution of Restaurants by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")

plt.show()
```
