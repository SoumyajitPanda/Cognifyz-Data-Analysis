Task: City Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Dataset
df = pd.read_csv("../Dataset .csv")
```

```python
# Highest number of restaurants
city_count = df["City"].value_counts()
```

```python
print("City with Highest Number of Restaurants")
print(city_count.head(1))
```

```python
# Average rating of each city
city_rating = df.groupby("City")["Aggregate rating"].mean().sort_values(ascending=False)
print("\nAverage Rating of Each City")
print(city_rating)
```

```python
# Highest average rating
print("\nCity with Highest Average Rating")
print(city_rating.head(1))
```

```python
# Visualization
top10 = city_rating.head(10)

plt.figure(figsize=(10,5))
plt.bar(top10.index, top10.values)

plt.xticks(rotation=90)
plt.title("Top 10 Cities by Average Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")

plt.show()
```
