```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Dataset
df = pd.read_csv("../Dataset .csv")
```

```python
# Count Online Delivery
online_delivery = df["Has Online delivery"].value_counts()
```

```python
# Percentage Calculation
percentage = (online_delivery / len(df)) * 100
```

```python
# Result Table
result = pd.DataFrame({
    "Online Delivery": online_delivery.index,
    "Restaurant Count": online_delivery.values,
    "Percentage (%)": percentage.values
})

result["Percentage (%)"] = result["Percentage (%)"].round(2)

print(result)
```

```python
# Average Rating Comparison
average_rating = df.groupby("Has Online delivery")["Aggregate rating"].mean()

print("\nAverage Ratings:")
print(average_rating)
```

```python
# Bar Chart - Restaurant Count
plt.figure(figsize=(6,4))
plt.bar(result["Online Delivery"], result["Restaurant Count"])
plt.title("Restaurants Offering Online Delivery")
plt.xlabel("Online Delivery")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
# Bar Chart - Average Rating
plt.figure(figsize=(6,4))
plt.bar(average_rating.index, average_rating.values)
plt.title("Average Rating by Online Delivery")
plt.xlabel("Online Delivery")
plt.ylabel("Average Rating")
plt.show()
```
